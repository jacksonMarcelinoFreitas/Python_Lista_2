```python
#Faça um programa que peça dois números, base e expoente, calcule e mostre o 
#primeiro número elevado ao segundo número. Não utilize a função de potência da linguagem.
base = (int(input("Informe a base da potencia: ")))
expoente = (int(input("Informe o expoente da potencia: ")))
potencia = 1 

for contador in range(expoente):
    potencia = potencia * base
    contador += 1

print("A potência é : %d" %(potencia))
```
